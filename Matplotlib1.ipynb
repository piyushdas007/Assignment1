```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# from a pandas dataset
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lables title
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue')

plt.title('Rohit Sharma Vs Virat Kolhi Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue',linestyle= 'dashdot')

plt.title('Rohit Sharma Vs Virat Kolhi Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashdot',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue',linestyle= 'dashdot',marker='o',markersize=10)

plt.title('Rohit Sharma Vs Virat Kolhi Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashdot',marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'blue',linestyle= 'dashdot',marker='o',markersize=10,label='Rohit')

plt.title('Rohit Sharma Vs Virat Kolhi Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# to label the line
plt.legend()
# plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
# plt.ylim(0,100000)
```

### Bar Chat

```python
# simple bar chat
children= [10,20,40,10,30]
colors=['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# bar chat using data
```

```python
# horizontal bar chat -> plt.barh
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem
children= [10,20,40,10,30]
colors=['red red red red ','blue blue blue blue','green green green green','yellow yellow yellow yellow','pink pink pink pink']

plt.bar(colors,children,color='black')
```

```python
children= [10,20,40,10,30]
colors=['red red red red ','blue blue blue blue','green green green green','yellow yellow yellow yellow','pink pink pink pink']

plt.bar(colors,children,color='black')
plt.xticks(rotation = 'vertical')
```

```python
# stacked bar chat
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')

plt.legend()
plt.show()
```

### Pie Chat

```python
# simple data
data = [20,45,75,84,100]
subjects = ['eng','science','maths','sst','hindi']

plt.pie(data,labels=subjects)

plt.show()
```
