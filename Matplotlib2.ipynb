```python
# A diff way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Stick Rate')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
# fig, ax = plt.subplots()

ax.scatter(batters['avg'],batters['strike_rate'], color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(18,9)) # sharex will share x axis

ax[0].scatter(batters['avg'],batters['strike_rate'],color= 'red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strikr Rate')
ax[0].set_ylabel('Strike Rate')
# ax[0].set_xlabel('Avg')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10)) # sharex will share x axis

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color= 'red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

ax[0,0].set_title('Avg Vs Strikr Rate')
ax[0,0].set_ylabel('Strike Rate')
# ax[0].set_xlabel('Avg')

ax[0,1].set_title('Avg Vs Runs')
ax[0,1].set_ylabel('Runs')
# ax[1].set_xlabel('Avg')
```

```python
fig = plt.figure()

ax1= fig.add_subplot(2,1,1) # (rows,cols,1st)
ax1.scatter(batters['avg'],batters['strike_rate'],color= 'red')

ax2 = fig.add_subplot(2,1,2) # (rows,cols,2nd)
ax2.hist(batters['avg'])
```

# 3D Surface Plot

```python
x = np.linspace(-10,10,100) # creating 100 points
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z, cmap='viridis')
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

# Contour Plot

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap='viridis') # contourf
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

# Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
